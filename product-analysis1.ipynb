```python
query = """
           SELECT customerNumber,priceEach,buyPrice FROM orderdetails od
           INNER JOIN orders o
           on o.orderNumber = od.orderNumber
           JOIN products p
           on p.productCode = od.productCode
"""
pd.read_sql(query,conn).head(10)
```

```python
query = """
           SELECT customerNumber,(priceEach-buyPrice) prof_margin 
           FROM orderdetails od
           INNER JOIN orders o
           on o.orderNumber = od.orderNumber
           JOIN products p
           on p.productCode = od.productCode
           limit 5
"""
pd.read_sql(query,conn).head(10)
```

```python
query = """
           SELECT customerNumber,(priceEach-buyPrice) prof_margin, quantityOrdered
           FROM orderdetails od
           INNER JOIN orders o
           on o.orderNumber = od.orderNumber
           JOIN products p
           on p.productCode = od.productCode
"""
pd.read_sql(query,conn).head(10)
```

```python
query = """
           SELECT customerNumber,((priceEach-buyPrice) * quantityOrdered) total_profit
           FROM orderdetails od
           INNER JOIN orders o
           on o.orderNumber = od.orderNumber
           JOIN products p
           on p.productCode = od.productCode
"""
pd.read_sql(query,conn).head(10)
```

```python
query = """
           SELECT customerNumber,SUM((priceEach-buyPrice) * quantityOrdered) total_profit
           FROM orderdetails od
           INNER JOIN orders o
           on o.orderNumber = od.orderNumber
           JOIN products p
           on p.productCode = od.productCode
           GROUP BY 1
           ORDER BY 2 DESC
"""
pd.read_sql(query,conn)
```

```python
query = """WITH
           Customer_profit As(
           SELECT customerNumber,SUM((priceEach-buyPrice) * quantityOrdered) total_profit
           FROM orderdetails od
           INNER JOIN orders o
           on o.orderNumber = od.orderNumber
           JOIN products p
           on p.productCode = od.productCode
           GROUP BY 1
           ORDER BY 2 DESC
           )
           SELECT customerNumber FROM Customer_profit
"""
           
pd.read_sql(query,conn)
```
